# Análisis Exploratorio de Datos (EDA)
Dataset: vehicles_us.csv

En este notebook se realiza un análisis exploratorio básico del conjunto de datos de anuncios de venta de coches en Estados Unidos.

```python
import pandas as pd
import plotly.express as px
```

```python
car_data = pd.read_csv("c:/programacion/triple10/proyecto_sprint7_vehicles/vehicles_us.csv")
car_data.head()
```

```python
car_data.info()
```

## Limpieza mínima de datos

Se eliminan filas con valores faltantes en columnas necesarias para el análisis gráfico:
- odometer
- model_year


```python
# Copiamos el dataset original para no modificarlo directamente
clean_data = car_data.copy()

# 1. Eliminamos filas sin kilometraje (necesario para histograma y scatter)
clean_data = clean_data.dropna(subset=["odometer"])

# 2. Eliminamos filas sin año del modelo
clean_data = clean_data.dropna(subset=["model_year"])

# 3. Convertimos model_year a entero
clean_data["model_year"] = clean_data["model_year"].astype(int)

# 4. Convertimos is_4wd a booleano (1 = True, NaN = False)
clean_data["is_4wd"] = clean_data["is_4wd"].fillna(0).astype(bool)

# Verificamos resultado
clean_data.info()
```

```python
fig = px.histogram(
    clean_data,
    x="odometer",
    title="Distribución del kilometraje",
)

fig.show()
```

```python
fig = px.scatter(
    clean_data,
    x="odometer",
    y="price",
    opacity=0.5,
    title="Relación entre kilometraje y precio"
)

fig.show()
```

```python
print("Filas originales:", car_data.shape[0])
print("Filas después de limpieza:", clean_data.shape[0])
print("Filas eliminadas:", car_data.shape[0] - clean_data.shape[0])
```

## Conclusiones del Análisis Exploratorio de Datos (EDA)

### 1. Limpieza de datos

El conjunto de datos original contenía 51,525 registros.  
Después de eliminar filas con valores faltantes en las columnas `odometer` y `model_year`, quedaron 40,563 registros válidos para el análisis.

Se eliminaron 10,962 registros incompletos, lo que representa aproximadamente el 21% del dataset original.

La limpieza fue mínima y justificada, ya que estas variables son necesarias para construir visualizaciones confiables.

---

### 2. Distribución del kilometraje (Histograma)

El histograma muestra una distribución claramente sesgada hacia la derecha:

- La mayoría de los vehículos se concentran entre 0 y 200,000 millas.
- Existe una cola larga que se extiende hasta aproximadamente 1,000,000 de millas.
- Se observan valores extremos (outliers) con kilometraje muy elevado.

Esto indica que la mayor parte de los autos en venta tienen un uso moderado, mientras que un pequeño grupo presenta un desgaste significativamente mayor.

---

### 3. Relación entre kilometraje y precio (Gráfico de dispersión)

El gráfico de dispersión muestra una relación negativa clara entre el kilometraje y el precio:

- A medida que aumenta el kilometraje, el precio tiende a disminuir.
- Los vehículos con menor kilometraje presentan mayor variabilidad en precios.
- Se observan algunos outliers con precios altos y kilometrajes elevados, así como vehículos extremadamente baratos.

Esta tendencia es consistente con el comportamiento esperado del mercado automotriz: el desgaste acumulado impacta directamente el valor del vehículo.

---

### 4. Observaciones generales

- La relación entre precio y kilometraje es visualmente significativa.
- Existen valores extremos tanto en kilometraje como en precio que podrían influir en análisis más avanzados.
- La limpieza aplicada permitió obtener visualizaciones más coherentes y representativas del comportamiento real de los datos.

En conclusión, el dataset muestra patrones consistentes con el mercado de autos usados y está en condiciones adecuadas para integrarse en la aplicación web desarrollada con Streamlit.
